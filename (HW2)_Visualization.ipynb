```python
#Import các thư viện cần thiết

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings; warnings.filterwarnings(action='once')
import random
import math
```

```python
#Load dữ liệu từ file
newdf = pd.read_csv("batdongsan.com.vn.csv", dtype = {"month": "string" , "id" : "string"})
print("DONE")
```

```python
# Tạo cột Price/m^2:
def price_per_square(row):
    return row['price'] / row['square']

newdf['pricePerSquare'] = newdf.apply(price_per_square, axis=1)
```

2. Cho đoạn code minh họa Biểu đồ thể hiện số bài đăng theo tháng của năm 2020. Hãy thực hiện các yêu cầu cho các câu hỏi dưới đây:

```python
df1 = newdf.groupby('month').size().reset_index(name='number of records')
n = df1['month'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi= 80)
plt.bar(df1['month'], df1['number of records'], color=c, width=.5)
for i, val in enumerate(df1['number of records'].values):
    plt.text(i, val, float(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12})

plt.gca().set_xticklabels(df1['month'], rotation=60, horizontalalignment= 'right')
plt.title("Biểu đồ thể hiện số bài đăng theo tháng của năm 2020", fontsize=22)
plt.ylabel('number of records')
plt.show()
```

* Biểu đồ thể hiện giá nhà đất trung bình theo tháng (VND/m2)

```python
df1 = newdf.groupby('month')['pricePerSquare'].mean().reset_index(name='the avg price')
n = df1['month'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi= 80)
plt.bar(df1['month'], df1['the avg price'], color=c, width=.5)
for i, val in enumerate(df1['the avg price'].values):
    plt.text(i, val, f"{val:.2f}", horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12})

plt.gca().set_xticklabels(df1['month'], rotation=60, horizontalalignment= 'right')
plt.title("Biểu đồ thể hiện giá nhà đất trung bình theo tháng(VND/m2)", fontsize=22)
plt.ylabel('the avg price')
plt.show()
```

# * Biểu đồ mật độ giá nhà của tháng 4 và tháng 12 (VND/m2)

```python
#Gọi ý: sns.displot()

# Vẽ biểu đồ mật độ giá nhà của tháng 4 và tháng 12
# Giả sử dataframe có cột 'pricePerSquare' và 'month'

newdf_april = newdf[newdf['month'] == '04']
newdf_december = newdf[newdf['month'] == '12']

plt.figure(figsize=(11,6))

# Sử dụng histplot với KDE để biểu diễn cả phân phối và biểu đồ mật độ
sns.histplot(newdf_april['pricePerSquare'], kde=True, color='orange', label='Tháng 4', stat="density", bins=20)
sns.histplot(newdf_december['pricePerSquare'], kde=True, color='blue', label='Tháng 12', stat="density", bins=20)

# Tùy chỉnh biểu đồ
plt.title('Biểu đồ mật độ giá nhà của tháng 4 và tháng 12 (VND/m2)')
plt.xlabel('pricePerSquare')
plt.ylabel('Density')
plt.legend()

# Hiển thị biểu đồ
plt.show()
```

3. Tương quan giữa các trường thuộc tính

* https://vi.wikipedia.org/wiki/H%E1%BB%87_s%E1%BB%91_t%C6%B0%C6%A1ng_quan
* bieu do the hien he so tuong quan giua cac cap bien so thuc banggf biểu đồ heatmap

```python
# Gợi ý: sns.heatmap(cmap='RdYlGn', ...)
# Biểu đồ thể hiện correlation giữa các thuộc tính
corr_matrix = newdf[['price', 'square', 'bedrooms', 'bathrooms']].corr()

# Vẽ heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='RdYlGn', vmin=-1, vmax=1)
plt.title('Heatmap hệ số tương quan giữa các biến số thực')
plt.show()
```

* bieu do the hien moi quan he giua cac bien so thuc với biểu đồ phân tán

```python
#Gợi ý: sns.pairplot(kind='scatter', ...)

sns.pairplot(newdf[['price', 'square', 'bedrooms', 'bathrooms']], kind='scatter')
plt.suptitle('Biểu đồ phân tán giữa các biến số thực', y=1.02)
plt.show()
```

4. Mật độ giá nhà trung bình theo quận

```python
#Gợi ý: sns.displot()
df_price_by_district = newdf.groupby('district')['pricePerSquare'].mean().reset_index()

# Vẽ biểu đồ mật độ giá nhà trung bình theo quận
plt.figure(figsize=(10, 6))
sns.displot(df_price_by_district['pricePerSquare'], kde=True)
plt.title('Mật độ giá nhà trung bình theo quận (VND/m2)')
plt.xlabel('Giá nhà trung bình (VND/m2)')
plt.ylabel('Density')
plt.show()
```
