```python
df = pd.read_csv('kc_house_data.csv')

# Tiền xử lý cột date
df['date'] = pd.to_datetime(df['date'])
df['Years'] = df['date'].dt.year

X = df.drop(columns=['id', 'date', 'price', 'waterfront', 'view', 'condition'])  # Loại bỏ cột không cần chuẩn hóa
y = df['price']

# Chuẩn hóa các dữ liệu dạng numerical
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

df_scaled = pd.concat([X_scaled, df['Years'], df['view'], df['waterfront'], df['condition']], axis=1)


# Chia data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_scaled, y, test_size=0.2, random_state=42)

print(df_scaled)

```

```python
# Build model
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Đầu tiên là LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)

# Giờ ta thử với Ridge, Lasso

from sklearn.linear_model import Ridge, Lasso

# Ridge
param_grid = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100]}
grid_search_ridge = GridSearchCV(Ridge(), param_grid=param_grid, cv=5, scoring = 'r2')
grid_search_ridge.fit(X_train, y_train)

y_pred = grid_search_ridge.predict(X_test)

r2 = r2_score(y_test, y_pred)
print(f"R-squared (Ridge): {r2}")

# Lasso
param_grid = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100]}
grid_search_lasso = GridSearchCV(Lasso(), param_grid=param_grid, cv=5, scoring = 'r2')
grid_search_lasso.fit(X_train, y_train)

y_pred = grid_search_lasso.predict(X_test)
r2 = r2_score(y_test, y_pred)
print(f"R-squared (Lasso): {r2}")

# Giờ ta thử KNN

from sklearn.neighbors import KNeighborsRegressor
param_grid = {'n_neighbors': [3, 5, 7, 9, 11]}
grid_search_knn = GridSearchCV(KNeighborsRegressor(), param_grid=param_grid, cv=5, scoring = 'r2')
grid_search_knn.fit(X_train, y_train)

y_pred = grid_search_knn.predict(X_test)

r2 = r2_score(y_test, y_pred)
print(f"R-squared (KNN): {r2}")

# Cuối cùng là RandomForrest

from sklearn.ensemble import RandomForestRegressor
param_grid = [
   {'n_estimators' : [30, 50],
    'max_depth': [10, 30], 'random_state': [42]}
]
grid_search_rf = GridSearchCV(RandomForestRegressor(), param_grid = param_grid, cv = 3, scoring = "r2")
grid_search_rf.fit(X_train, y_train)


y_pred = grid_search_rf.predict(X_test)

r2 = r2_score(y_test, y_pred)
print(f"R-squared (Random Forest): {r2}")


```
